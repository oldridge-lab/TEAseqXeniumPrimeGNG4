# This notebook contains neighborhood analysis 

Domain generation is done using L1 annotation

Enrichment analysis is done using L2 annotation

(original scripts adapted from https://github.com/nolanlab/NeighborhoodCoordination)

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
import time
import sys
import os

from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import OneHotEncoder

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

### Fill In:
    -ks = individual window sizes to collect.  
        -This accelerates the process of trying different window sizes by collecting them all in one step.
    -path_to_data:  file path for csv file with x,y cluster information
    -X:  column to use for X location
    -Y:  column to use for Y location
    -reg:  column to use for region (should be different for every tissue)
    -file_type('csv','pickle'):  file type
    -cluster_col : column name of cell type cluster to be used for neighborhoods
    -cellhierpath:  path to cellhier libary (should only have to set once per computer
    -keep_cols:  columns to transfer from original dataframe to dataframe with windows and neighborhoods

```python
ks = [60,80,100] # k=5 means it collects 5 nearest neighbors for each center cell
path_to_data = '/mnt/isilon/oldridge_lab/irinaz/xenium_tonsil/neighborhood/06302025/cell_annotation_df.csv' #annotation generated in 06302025

# data loaded form image (x = y; y = x - seurat take into account of this while creating spatial plots)
X = 'x'
Y = 'y'
reg = 'sample'
file_type = 'csv'

l1_cluster_col = 'l1_cluster_id'
l2_cluster_col = 'l2_cluster_id'
keep_cols = [X,Y,reg,l1_cluster_col,l2_cluster_col]

save_path = ''
os.makedirs(save_path,exist_ok=True)
```

```python
# read in cell_annotation_df (csv or pickle) and do some quick data rearrangement

n_neighbors = max(ks)
assert (file_type=='csv' or file_type =='pickle') #

if file_type == 'pickle':
    cells = pd.read_pickle(path_to_data)
if file_type == 'csv':
    cells = pd.read_csv(path_to_data)

# data loaded form image (x = y; y = x - seurat take into account of this while creating spatial plots)
cells.rename(columns={'x': '_tmp', 'y': 'x'}, inplace=True)
cells.rename(columns={'_tmp': 'y'}, inplace=True)

# add one hot encoding for each cluster
cells = pd.concat([cells,pd.get_dummies(cells[l1_cluster_col])],1) 

#cells = cells.reset_index() #Uncomment this line if you do any subsetting of dataframe such as removing dirt etc or will throw error at end of next next code block (cell 6)

sum_cols = cells[l1_cluster_col].unique() # vecor of cell annotation (0-25)
values = cells[sum_cols].values # (cell, 26)

print(cells.head(20))

# cell type (0-25)
print(sum_cols)

# one hot cell annotation
print(values.shape)
print(values)
```

```python
#find windows for each cell in each tissue region

def get_windows(job,n_neighbors):
    '''
    For each region and each individual cell in dataset, return the indices of the nearest neighbors.
    
    'job:  meta data containing the start time,index of region, region name, indices of region in original dataframe
    n_neighbors:  the number of neighbors to find for each cell
    '''
    start_time,idx,tissue_name,indices = job
    job_start = time.time()
    
    print ("Starting:", str(idx+1)+'/'+str(len(exps)),': ' + exps[idx])

    tissue = tissue_group.get_group(tissue_name)
    to_fit = tissue.loc[indices][[X,Y]].values

    #   fit = NearestNeighbors(n_neighbors=n_neighbors+1).fit(tissue[[X,Y]].values)
    fit = NearestNeighbors(n_neighbors=n_neighbors).fit(tissue[[X,Y]].values)
    m = fit.kneighbors(to_fit)
    #     m = m[0][:,1:], m[1][:,1:]
    m = m[0], m[1]
    
    #sort_neighbors
    args = m[0].argsort(axis = 1)
    add = np.arange(m[1].shape[0])*m[1].shape[1]
    sorted_indices = m[1].flatten()[args+add[:,None]]

    neighbors = tissue.index.values[sorted_indices]
   
    end_time = time.time()
   
    print ("Finishing:", str(idx+1)+"/"+str(len(exps)),": "+ exps[idx],end_time-job_start,end_time-start_time)
    
    return neighbors.astype(np.int32)


tissue_group = cells[[X,Y,reg]].groupby(reg) # will have a dict_keys of each sample ID, and then for each  [x,y,sample]

# Grab the list of unique tissue IDs
exps = list(cells[reg].unique())

# build a job tuple for each tissue/sample
tissue_chunks = [(time.time(),exps.index(t),t,a) for t,indices in tissue_group.groups.items() for a in np.array_split(indices,1)] 

# call each sample for get_window function - return indices of nearest neighbors
tissues = [get_windows(job,n_neighbors) for job in tissue_chunks] # tissue[0][2] - sample 0 and window - (,50)
```

```python
#for each cell and its nearest neighbors, reshape and count the number of each cell type in those neighbors
out_dict = {}
for k in ks:
    for neighbors,job in zip(tissues,tissue_chunks):

        chunk = np.arange(len(neighbors))#indices
        tissue_name = job[2]
        indices = job[3]
        window = values[neighbors[chunk,:k].flatten()].reshape(len(chunk),k,len(sum_cols)).sum(axis = 1)
        out_dict[(tissue_name,k)] = (window.astype(np.float16),indices)

#concatenate the summed windows and combine into one dataframe for each window size tested.
windows = {}
for k in ks:
   
    window = pd.concat([pd.DataFrame(out_dict[(exp,k)][0],index = out_dict[(exp,k)][1].astype(int),columns = sum_cols) for exp in exps],0)
    window = window.loc[cells.index.values]
    window = pd.concat([cells[keep_cols],window],1)
    windows[k] = window
    
print(windows) # coordinates, cell types, neighborhood cell types (count for each clusters) - for each window size i.e. window[40]
```

### Fill In:
    -k:  window_size (needs to be in ks (saved above))
    -n_neighborhoods:  number of neighborhoods to find

```python
k = 60                #windows (neighbor to find)
n_neighborhoods = 18  #clusters
neighborhood_name = "neighborhood"+"k"+str(k)+"n"+str(n_neighborhoods)
k_centroids = {}
```

```python
windows2 = windows[k] # k
#windows2[cluster_col] = cells[cluster_col]

km = MiniBatchKMeans(n_clusters = n_neighborhoods,random_state=0)

labelskm = km.fit_predict(windows2[sum_cols].values)
k_centroids[k] = km.cluster_centers_

cells[neighborhood_name] = labelskm

cells[neighborhood_name] = cells[neighborhood_name].astype('category')
```

```python
# save csv file
print(cells)
cells.to_csv("./neighborhoods.csv", index=False)
```

```python
# read the annotation/cell label CSV written from R

path_to_label = ''
label_df = pd.read_csv(path_to_label)          # columns: cluster_id, cluster

label_map = dict(zip(label_df['cluster_id'], label_df['l2_cluster']))

cell_order        = label_df['cluster_id'].tolist()          # [0, 1, 2, …]
cell_order_labels = [label_map[i] for i in cell_order]       # ['Mono/Mac', …]

print(label_map)   # 'Mono/Mac'
print(cell_order)
```

```python
# generate one-hot encoding for l2
cell_labels_l2 = cells['l2_cluster_id'].values.reshape(-1, 1)

enc = OneHotEncoder(sparse=False, dtype=int)
values = enc.fit_transform(cell_labels_l2)  # shape = (n_cells, n_l2_clusters)
#print(values.shape)

# get niche composition based on l2 analysis
# neighborhood_ids = np.sort(cells[neighborhood_name].astype(int).unique())
# Convert the *column* itself to int once so everything downstream is numeric
cells[neighborhood_name] = cells[neighborhood_name].astype(int)
neighborhood_ids = np.sort(cells[neighborhood_name].unique())  # array([0..9])
cells[neighborhood_name] = cells[neighborhood_name].astype('category')
print(neighborhood_ids)

n_niches = len(neighborhood_ids) #10
n_ctypes = values.shape[1] #41

# Compute niche_clusters: average composition for each niche
niche_clusters = np.zeros((n_niches, n_ctypes))

for idx, niche in enumerate(neighborhood_ids):
    niche_idx = np.where(cells[neighborhood_name] == niche)[0]
    niche_clusters[idx, :] = values[niche_idx].mean(axis=0)

tissue_avgs = values.mean(axis=0)  # shape = (n_ctypes,)

# enrichment analysis
fc = np.log2(
        ((niche_clusters + tissue_avgs) /
         (niche_clusters + tissue_avgs).sum(axis=1, keepdims=True))
        / tissue_avgs
     )

cell_order_labels = [label_map[i] for i in cell_order]


fc_df = pd.DataFrame(fc, columns=cell_order_labels, index=neighborhood_ids)
fc_df = fc_df[cell_order_labels]  # Reorder explicitly if desired
#print(fc_df)

s = sns.clustermap(
    fc_df,
    vmin=-3, vmax=3,
    cmap='bwr',
    row_cluster=False,
    col_cluster=False, 
    xticklabels=cell_order_labels
)

# save the figure
folder_path = ""
os.makedirs(folder_path, exist_ok=True)
outpath = os.path.join(folder_path, "enrichment_plots.png")
plt.savefig(outpath)
```

```python
# Save Figures

# ensure neighborhood is categorical
cells[f'neighborhoodk{k}n{n_neighborhoods}'] = cells[f'neighborhoodk{k}n{n_neighborhoods}'].astype('category')

# sort list of samples
samples = sorted(cells['sample'].unique())
categories = cells[f'neighborhoodk{k}n{n_neighborhoods}'].cat.categories
n_categories = len(categories)

# colormap
cmap = plt.get_cmap('Set3', n_categories)

for sample in samples:
    df = cells[cells['sample'] == sample]
    
    fig, ax = plt.subplots(figsize=(8, 8), dpi=300)
    sc = ax.scatter(
        df['x'], df['y'],
        c=df[f'neighborhoodk{k}n{n_neighborhoods}'].cat.codes,
        cmap=cmap, vmin=0, vmax=n_categories-1,
        s=1, marker='.'
    )
    
    ax.set_title(sample, fontsize=16)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # legend
    handles = [
        Line2D([0], [0], marker='o', color=cmap(i),
               linestyle='', markersize=6)
        for i in range(n_categories)
    ]
    ax.legend(
        handles, categories,
        title=f'neighborhoodk{k}n{n_neighborhoods}',
        loc='center left',
        bbox_to_anchor=(1, 0.5),
        fontsize='small',
        title_fontsize='small',
        frameon=False
    )
    
    fig.tight_layout(rect=[0, 0, 0.8, 1])
    
    # save
    folder_path = ""
    os.makedirs(folder_path, exist_ok=True)

    outpath = os.path.join(folder_path, f"{sample}.png")
    fig.savefig(outpath,dpi=300)
    
    plt.close(fig) 
    print(f"→ Saved {outpath}")
```
